```java
//Crear una clase para hacer una abstraccion de un gato

class Cat {
    //atributos/propiedades/caracteristicas(definen el estado de un objeto)
    String nombre;
    String color;
    String raza;
    String tamaño;
    int edad;
    Color color1;
    
    //metodos/funciones(acciones que ese objeto puede hacer)(definen el comportamiento de un objeto)
    void speak(){ 
        println("Miauuuuu...");
    }
    void printCat(){ 
        printf("Nombre: %s %n",this.nombre);
        printf("raza: %s %n",this.raza);
        //printf("color: %s %n",this.color1);
    }
    public String toString(){
          return "El gato " + this.nombre + " tiene el color " + this.color1.toString();
    }
}
//nueva instancia 
Cat gato1=new Cat();
Cat gato2=new Cat();
gato1.color1=new Color();
gato2.color1=new Color();
//accedemos a la instancia y asignamos valores
gato1.nombre="Rodolfo";
gato1.raza="siames";
gato2.nombre="Jacinto";
gato1.raza="sphynx";

gato1.color1.name="coral";
gato1.color1.red=256;
gato1.color1.blue=22;
gato1.color1.yellow=345;

gato2.color1.name="blue";
gato2.color1.red=254;
gato2.color1.blue=23;
gato2.color1.yellow=2;

//printamos 
gato1.printCat();
gato1.speak();
println(gato1.toString());
println(gato2.toString());

//si queremos printar todos los metemos en un array

Cat[]gatos={gato1,gato2};
for(Cat gato:gatos){
    println("-------------------");
    gato.printCat();
    gato.speak();
    println(gato1.toString());
}
```

```java
class Car{
    Color color1;
    String marca;
    String tamaño;
    String color;
    String state = "good";
    
    void printCar(){
        printf("color: %s %n", this.color1);
        printf("marca: %s %n", this.marca);
        printf("tamaño: %s %n", this.tamaño);
    }
    void acelerar(){
        printf("%s hace ruuuumm... %n", this.marca);
    }
    void frenar(){
        printf("%s hace grrrrrr... %n", this.marca);
    }
    
    String sound(){
        if(this.state == "good"){
            return "Bruuuuumm";
        }else{
            return "Crick crick";
        }
    }
    
    public String toString(){
        return "El coche " + this.marca + " tiene el color " + this.color1.toString() 
        + " y hace el sonido: " + this.sound() ;
    }
}

Car coche1 = new Car();
Car coche2 = new Car();
//CREAMOS UNA INSTANCIA PARA COCHE1
coche1.color1=new Color();
//ACCEDEMOS A LOS VALORES DEL COLOR DEFINIDO ABAJO
coche1.color1.name="Coral";
coche1.color1.red=256;
coche1.color1.blue=22;
coche1.color1.yellow=345;
//dar un valor al color del coche coche1:
coche1.marca="Mercedes";
coche1.tamaño="mediano";
coche2.color="blanco";
coche2.marca="Audi";
coche2.tamaño="grande";

coche1.printCar();
coche1.acelerar();
coche2.frenar();

coche1.toString();
```

```java
class Color{
    int red;
    int blue;
    int yellow;
    String name;

    public String toString(){
        return this.name + " rgb:" + this.red + "," + this.blue + "," + this.yellow;
    }
}

Color color1 = new Color();
color1.red=256;
color1.name="coral";
color1.blue=22;
color1.yellow=345;

color1.toString();
```

```java
class Rectangle {

    double base;
    double altura;
    Color backgroundColor;
    Color lineColor;
    
    double area(){
        return this.base*this.altura;
    }
    double perimeter(){
        return 2 * this.base + 2 * this.altura;
    }
    
    public String toString(){
        return "las base del rectangulo es: " + this.base + "\nla altura es: " + this.altura 
            + "\nel area del rectangulo es: " + this.area() + "\nel perimetro del rectangulo es: " 
            + this.perimeter() + "\nel backgroundColor es: " + this.backgroundColor.toString() 
            + "\nel lineColor es: " + this.lineColor.toString();
    }
}

Rectangle rectangulo = new Rectangle();
rectangulo.base = 10;
rectangulo.altura = 20;

Color colorDeFondo = new Color();
Color colorDeLinea = new Color();

colorDeFondo.name="blanco";
colorDeFondo.red=234;
colorDeFondo.blue=20;
colorDeFondo.yellow=345;

colorDeLinea.name="negro";
colorDeLinea.red=210;
colorDeLinea.blue=240;
colorDeLinea.yellow=245;

rectangulo.backgroundColor=colorDeFondo;
rectangulo.lineColor=colorDeLinea;

println(rectangulo.toString());
```
